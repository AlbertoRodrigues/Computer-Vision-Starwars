## Let's implement some functions on images and use functions from the opencv library

### Grayscale image transformation

```python
def gray(img):    
    return(np.floor(0.114*img[:,:,0]+0.587*img[:,:,1]+0.299*img[:,:,2]).astype("uint8"))
```

```python
Image.fromarray(gray(img))
```

```python
gray2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
Image.fromarray(gray2)
```

### The opencv library uses the BGR channel pattern

### Flipping an image

```python
def flip(img):    
    return(img[:,::-1,:])
```

#### Original image

```python
Image.fromarray(img)
```

```python
Image.fromarray(flip(img))
```

```python
Image.fromarray(cv2.flip(img,1))
```

### Random zoom 

```python
tamanho_janela=np.int(np.ceil(img.shape[0]/2))
linha_zoom=np.random.randint(0, img.shape[0]-tamanho_janela)
coluna_zoom=np.random.randint(0, img.shape[1]-tamanho_janela)
img_zoom=img[linha_zoom:(linha_zoom+tamanho_janela),coluna_zoom:(coluna_zoom+tamanho_janela),:]
Image.fromarray(cv2.resize(img_zoom,(img.shape[0], img.shape[1])))
```

```python
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(300,300))
```

```python
Image.fromarray(dst)
```

### Rotating an image

```python
rotated = imutils.rotate_bound(img, 92)
Image.fromarray(rotated)
```

### Changing the brightness and contrast 

```python
def brightness_constrast(img, alpha, beta):
    new_image = np.zeros(img.shape, img.dtype)
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            for c in range(img.shape[2]):
                new_image[y,x,c] = np.clip(alpha*img[y,x,c] + beta, 0, 255)
    return(new_image)
```

```python
Image.fromarray(brightness_constrast(img,1,100))
```

```python
Image.fromarray(brightness_constrast(img,0.5,0))
```

```python
Image.fromarray(brightness_constrast(img,0.3,80))
```

### Usual filters

```python
Image.fromarray(img)
```

```python
def blur(img, tam):
    return(cv2.blur(img, (tam,tam) ))
```

```python
Image.fromarray(blur(img, 5))
```

```python
Image.fromarray(blur(img, 10))
```

```python
def gaussian_blur(img, tam):
    return(cv2.GaussianBlur(img, (tam,tam), 0 ))
```

```python
Image.fromarray(gaussian_blur(img, 5))
```

```python
Image.fromarray(blur(img, 10))
```

```python
def median(img, tam):
    return(cv2.medianBlur(img, tam))
```

```python
Image.fromarray(median(img, 5))
```

```python
Image.fromarray(median(img, 9))
```
